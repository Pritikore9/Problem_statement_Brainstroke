# 3 create new column by above category 0-15 -kid, 15-30 - teenage, 30-50 -adult, 50 & above Oldage

```python
data
```

```python
Age_cat=[]
for x in data["age"]:
    if x<15:
        Age_cat.append("kid")
    elif x>15 and x<30:
        Age_cat.append("teenage")
    elif x>30 and x<50:
        Age_cat.append("adult")
    else:
        Age_cat.append("oldage")
```

```python
data.insert(2,"Age_cat",Age_cat)
```

```python
data
```

```python
ans=data.iloc[:,[1,2]]
```
